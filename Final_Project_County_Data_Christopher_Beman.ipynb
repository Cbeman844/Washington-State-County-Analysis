```python
# Importing Required Libraries
import pandas as pd
```

```python
# Read in the data from a csv file
pd.set_option('display.max_rows', None)
license_data = pd.read_csv('license_data.csv')
license_data.head()
```

```python
# Gets the users choice of database and checks for validity
def getUserInputDataSet():
    dataset = input("Which dataset would you like to work with? (s = state, c = country) ")

    if dataset == "s":
        userDataSet = "state"
    elif dataset == "c":
        userDataSet = "country"
    else:
        print("Invalid input")

    return(userDataSet)
```

```python
# Gets the users choice of county and checks for validity
def getCountyInput():
    countyValid = False
    while countyValid == False:
        county = input("Which county whould you like to look at?")
        if (license_data['county'].eq(county)).any():
            countyValid = True
            print("You selected",county ,"County")
            print()
            return(county)
        else:
            print("Invalid county try capitalizing the first letter")
```

```python
# Shows the county data requires a valid county in the data
def showCounty(userCounty):

    king_county_countries = license_data.loc[(license_data.county == userCounty) & (license_data.prior_country == "USA")]

    # group the filtered data with counts of each country and sort by the count max  to lowest
    king_county_countries_group = king_county_countries.groupby("prior_state")["prior_state"].count().sort_values(ascending=False)
    # print the entire dataframe
    print(king_county_countries_group)
    print()
```

```python
# Gets the input from user for which state to look at with validating
def getStateInput():
    stateValid = False
    while stateValid == False:
        # Validity checking process matches input to database
        state = input("Which state whould you like to look at?")
        if (license_data['prior_state'].eq(state)).any():
            stateValid = True
            print("You selected the State of:", state)
            print()
            return(state)
        else:
            print("Invalid county try capitalizing the first letter")
```

```python
# Gets the Users input for a valid country
def getCountryInput():
    countryValid = False
    while countryValid == False:
         # Validity checking process matches input to database
        country = input("Which country whould you like to look at?")
        if (license_data['prior_country'].eq(country)).any():
            countryValid = True
            print("You selected:", country)
            print()
            return(country)
        else:
            print("Invalid county try capitalizing the first letter")    
```

```python
# Shows the data for a specific country
def showCountryData(country):
    county_countries = license_data.loc[(license_data.prior_country != "USA") & (license_data.prior_country == country)]
    countyCountryGroup = county_countries.groupby("prior_country")
    countyCountryGroup = countyCountryGroup.head(5)

    print(countyCountryGroup)
```

```python
# Shows the data for the top 5 countries into Washington
def top5FromAllCountry():
    print("These are the top 5 countries that people have moved from")
    countyCountriesGroup = license_data.groupby("prior_country")["prior_country"].count().sort_values(ascending=False)
    countyCountriesGroup = countyCountriesGroup.head(5)

    print(countyCountriesGroup)
    print()
```

```python
# Shows the top 5 countries for a specific county requires valid county input
def top5MovedFromCountry(county):
    print("These are the top 5 countries that people have moved to", county, "county")
    county_countries = license_data.loc[(license_data.county == county) & (license_data.prior_country != "USA")]
    # group the filtered data with counts of each country and sort by the count max  to lowest
    countyCountriesGroup = county_countries.groupby("prior_country")["prior_country"].count().sort_values(ascending=False)
    countyCountriesGroup = countyCountriesGroup.head(5)

    print(countyCountriesGroup)
    print()
```

```python
# Gets the percentage moved into a specific county based off the whole population
def percentageMovedToCountyCountry(county):
    allCountyCount = len(license_data)

    county_countries = license_data.loc[(license_data.county == county) & (license_data.prior_country != "USA")]
    countyCount = len(county_countries)
    
    print("Percentage of people who moved to", county, "county is from out of the country:", str(round(((countyCount/allCountyCount)*100)))+'%')
```

```python
# Gets the top counties other than the top 3
def mostPopularCountiesOtherCountry():

    print("Most popular counties to move to other than King, Snohomish, or Pierce")
    county_countries = license_data.loc[(license_data.county != "King") & (license_data.county != "Snohomish") & (license_data.county != "Pierce") & (license_data.card_type == "Driver License") & (license_data.prior_country != "USA")]
    countyCountriesGroup = county_countries.groupby("county")["county"].count().sort_values(ascending=False)
    countyCountriesGroup = countyCountriesGroup.head(3)
    
    print(countyCountriesGroup)
```

```python
# Gets the most and least likely month and year for Drivers license and ID cards
def monthAndYearLicenseIDCountries():
    county_countries = license_data.loc[(license_data.card_type == "ID Card") | (license_data.card_type == "Driver License") & (license_data.prior_country != "USA")]
    
    print("The most popular month for transferring licenses and IDs for out of country is (with number transferrd):")
    countyCountriesGroup = county_countries.groupby("month")["month"].count().sort_values(ascending=False)
    countyCountriesGroup = countyCountriesGroup.head(1)
    print(countyCountriesGroup)
    print()

    print("The least popular month for transferring licenses and IDs for out of country is (with number transferrd):")
    countyCountriesGroup = county_countries.groupby("month")["month"].count().sort_values(ascending=True)
    countyCountriesGroup = countyCountriesGroup.head(1)
    print(countyCountriesGroup)
    print()

    print("The most popular year for transferring licenses and IDs for out of country is (with number transferrd):")
    countyCountriesGroup = county_countries.groupby("year")["year"].count().sort_values(ascending=True)
    countyCountriesGroup = countyCountriesGroup.head(1)
    print(countyCountriesGroup)
    print()

    print("The least popular year for transferring licenses and IDs for out of country is (with number transferrd):")
    countyCountriesGroup = county_countries.groupby("year")["year"].count().sort_values(ascending=False)
    countyCountriesGroup = countyCountriesGroup.head(1)
    print(countyCountriesGroup)
    print()
        
```

```python
# User menu and functionality
def menu():
    userExit = False
    while userExit == False:
        print("Welcome to the King County Data Repository!")

        #Shows the number of people who moved to washington from another country or state
        userDataSet = getUserInputDataSet()
        
        if userDataSet == "country":

            userValid = False
            while userValid == False:
                print("Thank you for choosing the country dataset!")
                print("Please pick one of the following items")
                print("Menu Item (1) Show a Country's Data")
                print("Menu Item (2) Top 5 Countries Into a Specific County")
                print("Menu Item (3) Percentage Moved Into a Specific County")
                print("Menu Item (4) Top 5 moved from all countries")
                print("Menu Item (5) Most Popular Counties Other than the Top 3")
                print("Menu Item (6) Most and Least Popular Month and Year for Licensing")
                print()

                userMenuInput = int(input("Which Menu Item would you like to pick? \nPlease enter a number from 1 to 6:"))
                print("You Selected Menu Item:", userMenuInput)

                if userMenuInput == 1:
                    # Menu Item 1 Show a Country's Data
                    userCountry = getCountryInput()
                    showCountryData(userCountry)
                    userValid = True

                elif userMenuInput == 2:
                    # Menu Item 2 Top 5 Countries Into a Specific County
                    userCounty = getCountyInput()
                    top5MovedFromCountry(userCounty)
                    userValid = True

                elif userMenuInput == 3:
                    # Menu Item 3 Percentage Moved Into a Specific County
                    userCounty = getCountyInput()
                    percentageMovedToCountyCountry(userCounty)
                    userValid = True

                elif userMenuInput == 4:    
                    # Menu Item 4 Top 5 moved from all countries        
                    top5FromAllCountry()
                    userValid = True

                elif userMenuInput == 5:
                    # Menu Item 5 Most Popular Counties Other than the Top 3
                    mostPopularCountiesOtherCountry()
                    userValid = True

                elif userMenuInput == 6:
                    # Menu Item 6 Most and Least Popular Month and Year for Licensing
                    monthAndYearLicenseIDCountries()
                    userValid = True
                
                else:
                    print("Please enter a valid number")   
                    userValid = False
            


        elif userDataSet == "state":
            
            userValid = False
            while userValid == False:
                print("Thank you for choosing the country dataset!")
                print("Please pick one of the following items")
                print("Menu Item (1) Show a State's Data")
                print("Menu Item (2) Top 5 States Into a Specific County")
                print("Menu Item (3) Percentage Moved Into a Specific County")
                print("Menu Item (4) Top 5 moved from all countries")
                print("Menu Item (5) Most Popular Counties Other than the Top 3")
                print("Menu Item (6) Most and Least Popular Month and Year for Licensing")
                print()

                userMenuInput = int(input("Which Menu Item would you like to pick? \nPlease enter a number from 1 to 6:"))
                print("You Selected Menu Item:", userMenuInput)

                if userMenuInput == 1:
                    # Menu Item 1 Show a States Data
                    userState = getStateInput()
                    showStateData(userState)
                    userValid = True

                elif userMenuInput == 2:
                    # Menu Item 2 Top 5 States Into a Specific County
                    userCounty = getCountyInput()
                    top5MovedFromState(userCounty)
                    userValid = True

                elif userMenuInput == 3:
                    # Menu Item 3 Percentage Moved Into a Specific County
                    userCounty = getCountyInput()    
                    percentageMovedToCountyStates(userCounty)
                    userValid = True    

                elif userMenuInput == 4:
                    # Menu Item 4 Top 10 moved from all countries 
                    top5FromAllStates()
                    userValid = True

                elif userMenuInput == 5:
                    # Menu Item 5 Most Popular Counties Other than the Top 3
                    mostPopularCountiesOtherStates()
                    userValid = True

                elif userMenuInput == 6:
                    # Menu Item 6 Most and Least Popular Month and Year for Licensing
                    monthAndYearLicenseIDStates()
                    userValid = True

                else:
                    print("Please enter a valid number")   
                    userValid = False

             


        print()
        userExitInput = input("Would you like to exit enter: e \nIf you would like to go again enter anything else ")
        if userExitInput.lower() == 'e':
            userExit = True

        print("Thank you for using the Washington State Data Repository Program!")




```
